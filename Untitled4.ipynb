```python
import pandas as pd
import numpy as np
import torch
import re
```

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
df = pd.read_csv('ftd_last2submission.csv')
df.shape
```

```python
d ={}
result=pd.DataFrame(columns=["id", "target"])
for i in df["target"]:
    j=i.split(":")
    #print(re.search(", +(\d+)\]\)", j[0]).groups()[0])
    d = {'id': int(re.search("tensor\(\[( *\d+),", j[0]).groups()[0]), 'target': int(re.search("tensor\(\[( *\d+),", j[1]).groups()[0])}
    result = result.append(pd.Series(data=d), ignore_index=True )
    d = {'id': int(re.search(", +(\d+)\]\)", j[0]).groups()[0]), 'target': int(re.search(", +(\d+)\]\)", j[1]).groups()[0])}
    result = result.append(pd.Series(data=d), ignore_index=True )
    print(result.head(10))
result.describe()
```

```python
Y_pred=[]
#Y_pred.append([0, 1])
for i in test_df["id"]:
    t = result.loc[lambda df: df['id'] == i]
    #print("iter",i, t)
    #print(t.shape)
    if min(t.shape)>0:
        Y_pred.append(t["target"].iloc[0])
    else:
        print("index", i, "потерялся")
len(Y_pred)
```

```python
import time
submission = pd.DataFrame({"id": result["id"], "target": Y_pred }, dtype = "int")
fname = '' + time.ctime() + 'submission.csv'
submission.to_csv(fname, index=False, header="id,target")
```
