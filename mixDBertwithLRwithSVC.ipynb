```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

```python
svc = pd.read_csv("data2/subms/SVCdbu.csv")
svc.head()
```

```python
lr = pd.read_csv("LR_test_.csv")
lr.head()
```

```python
bert = pd.read_csv("bert.csv")
bert.head()
```

```python
new_res = []
for b, s, p0, p1 in zip(bert["target"], svc["target"], lr["0"], lr["1"]):
    if p1 > p0:
        t = 1 
    else:
        t = 0
    if s + t + b > 1:
        new_res.append(1)
    else:
        if p0 > 0.4 and p0 < 0.6:
            new_res.append(b)
        else: 
            new_res.append(b)
    print(b ,s, t, new_res[-1], p0, p1)
len(new_res)
```

```python
import time
submission = pd.DataFrame({"id": svc["id"], "target": new_res }, dtype = "int")
fname = '' + time.ctime() + 'submission.csv'
submission.to_csv(fname, index=False, header="id,target")
```
